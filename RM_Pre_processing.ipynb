**Manipulations with the Health Safety Environment (HSE) dataset.**

```python
hse = pd.read_csv("./Datasets/Cleaned_HSE.csv", index_col = 0)
hse.head(40)
```

```python
#Code to use for Ordinal Encoder
injury_type = hse[["IADC Injury Classification"]]
injury_type.head(10)

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
injury_type_encoded = ordinal_encoder.fit_transform(injury_type)
injury_type_encoded[:10]
```

```python
#Code to use for OneHot Encoder
from sklearn.preprocessing import OneHotEncoder

# ohe = OneHotEncoder(use_cat_names=True)
# ohe.fit_transform(X)

cat_encoder = OneHotEncoder(sparse=False, drop=None) #drop="first"
injury_type_1hot = cat_encoder.fit_transform(injury_type)
cat_encoder.get_feature_names()
```

```python
#Convert the ndarray to a dataframe and then add it into existing df
hse = pd.concat([hse, pd.DataFrame(injury_type_1hot)], axis=1)

#renaming one hot encoding columns 
hse = hse.rename(columns={0:"Evaluation Only", 1: "First Aid Incident", 
                          2: "Medical Treatment Only", 
                          3: "No IADC Injury Classification", 
                          4: "Restricted Work Incident"})

#drop column we hot encoded to not have multicollinearity issues
hse = hse.drop(columns={"IADC Injury Classification"})
hse.head(10)

```

```python
# create new dataframe with the columns we need
df_hse_year_month = hse.groupby(["Rig Name","year","year_month"]).size().reset_index()
df_hse_year_month = df_hse_year_month.rename(columns = {0:"RM_Events_per_month"})
df_hse_year_month = df_hse_year_month.sort_values(by=["Rig Name"])
df_hse_year_month = df_hse_year_month.reset_index(drop=True)
df_hse_year_month["year"] = df_hse_year_month["year"].astype(int)
df_hse_year_month


```

```python
demo = df_hse_year_month.copy()

demo =  hse.groupby(["Rig Name","year","year_month"], as_index=False).sum()
demo = demo.drop(columns={"Form ID","month","day"})
demo = demo.sort_values(by=["Rig Name"])
demo["year"] = demo["year"] = demo["year"].astype(int)

list_name=['Evaluation Only',
 'First Aid Incident',
 'Medical Treatment Only',
 'No IADC Injury Classification',
 'Restricted Work Incident']
demo['Total_No_Events']=demo.loc[:,list_name].sum(axis=1)

demo

```

```python
def get_count_injury(df, cat_column):   
    count_injury = df.groupby(["Rig","year","year_month"])[cat_column].sum()
    count_injury = pd.Series(count_injury).reset_index(drop = True)
    return count_injury

ab_year_month['Alpha'] = get_count_injury(ab,"Alpha")
ab_year_month['RM_Alpha'] = ab_year_month["Alpha"]/ab_year_month["No_of_Events"]
ab_year_month['Beta'] = get_count_injury(ab,"Beta")
ab_year_month['RM_Beta'] = ab_year_month["Beta"]/ab_year_month["No_of_Events"]
ab_year_month = ab_year_month.drop(columns={"No_of_Events"})

ab_year_month = ab_year_month.drop(columns={"Alpha","Beta"})
ab_year_month

```

```python
#Find the total number of each variable per month
rm_observation = rm.groupby(["Rig_Name","year","year_month"]).sum().reset_index()
rm_observation = rm_observation.drop(columns={"month","day"})

#Find the total number of each variable per month accounting the 120days
rm_observation["RM_Total_Observations_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_low_risk_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Low_Risk']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_medium_risk_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Medium_Risk']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_high_risk_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_High_Risk']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_unsafe_action_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Unsafe_Action']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_unsafe_condition_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Unsafe_Condition']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_unsafe_condition_corrected_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Unsafe_Condition_Corrected']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_action_taken_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Action_Taken']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation["RM_positive_observations_120d"]=(rm_observation
                              .groupby('Rig_Name')['Count_Positive_Observation']
                              .rolling(min_periods=3, window=3, closed= "left")
                              .sum().reset_index(0,drop=True))

rm_observation = rm_observation.drop(columns={"Count","Count_Low_Risk","Count_Medium_Risk",
                                             "Count_High_Risk","Count_Unsafe_Action",
                                              "Count_Unsafe_Condition_Corrected",
                                              "Count_Unsafe_Condition","Count_Action_Taken",
                                              "Count_Positive_Observation"})
rm_observation
```
